# CDC Tobacco Survey Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import and take a look at the data

```python
# Import the dataset
df_turkey_2012 = pd.read_stata('data/TU2012_Public_Use_3Jun2014.dta')
# What is the shape of the dataset
df_turkey_2012.shape
```

There are 9851 rows and 264 columns in total.

```python
# Take a look at the first 10 rows of the dataset
df_turkey_2012.head(10)
```

Each of the column names appears to correspond to a specific survey question that can be found in the same zipped file downloaded with the .dat file from the CDC's website. Each of the rows appears to correspond to one of the subjects used in the survey. There are therefore ~264 survey questions and ~9,850 subjects interviewed for the survey. Each subject is interviewed on questions concerning their household.

For example, the HH1 column appears to refer to the question: 'How many persons live in this household?'
Each of the rows corresponds to a quantitative value indicating how many people live in that specific household. 

### Basic Exploratory Data Analysis
We're going to select a few of the columns for analysis, and make some simple data visualizations.

#### Age, education, frequency of current smoking
Taking a look at age of respondent ['age'], and how often the respondent currently smokes tobacco ['B01'].

```python
# Variable we'll use for the dataset
df_age = df_turkey_2012[['age', 'B01']]
```

```python
#create a function for generating the swarm plot
def boxplot(ax, x, y, xlabel, ylabel, title, data):
    ax = sns.boxplot(x, y, data=data)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title,fontsize=16)
```

```python
fig, ax = plt.subplots(figsize=(10,7))
plt.style.use('ggplot')
boxplot(ax, 'B01','age','How often the respondent smokes',
        'Age of respondent', 'Respondent\'s Age versus Frequency of Smoking',df_age)
plt.show()
```

```python
#create a function for generating the plot
def swarmplot(ax, x, y, hue, xlabel, ylabel, title, data):
    ax = sns.swarmplot(x, y, hue=hue, data=data)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title,fontsize=16)
```

```python
fig, ax = plt.subplots(figsize=(18,7))
swarmplot(ax, 'B07','age',None,'Time after waking up','Age',
          'How soon after you wake up do you usually have your first smoke?',df_turkey_2012)
plt.show()
```
